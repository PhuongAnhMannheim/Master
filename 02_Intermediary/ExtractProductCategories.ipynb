```python
# System libraries
import logging, os, sys
import gzip
import re
import pandas as pd
from urllib.parse import urlparse
import tldextract
```

```python
# Input
product_path = '../Data/schema_Product.gz'

# Output
prodcatlog_path = '../Logs/prodCat.log'
prodcatout = '../Data/prodCat.txt'
prodwebsites = '../Data/products_sites.txt'

logger = logging.getLogger()
fhandler = logging.FileHandler(filename=prodcatlog_path, mode='a')
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
fhandler.setFormatter(formatter)
logger.addHandler(fhandler)
logger.setLevel(logging.DEBUG)
```

```python
taxo_pattern = re.compile("<http://schema.org/Product/category>", re.IGNORECASE)
split_pattern = re.compile("^(_:.*)\s(<http:\/\/schema\.org\/Product\/category>)\s(.*)\s<(.*)>\s\.$", re.IGNORECASE)

with gzip.open(product_path,"rt") as f:
    i = 0
    detected = 0
    skipped = 0
    not_taxo = 0
    inserted = 0
    for line in iter(f.readline, ""):
        i += 1
        if not taxo_pattern.search(line):
            not_taxo += 1
            continue
        #print(line)
        match = split_pattern.match(line)
        if match is None:
            skipped += 1
            continue
        detected += 1
        props = match.groups()
        subj = props[0]
        pred = props[1]
        obj = props[2].lower
        source = props[3]
        with open(prodcatout,'a') as file:
            file.write( obj + "\n")
            inserted += 1

logging.debug("detected " + str(detected) + ", inserted: " + str(inserted) +" lines out of " + str(i) + "; not in taxo: " + str(not_taxo))
logging.debug("Done processing the product/categories file")
```

```python
data =[]
with open(prodcatout,'r') as file:
    for line in file:
        data.append(line.strip())
df = pd.DataFrame({'prodcat' :data})
df.describe(include='all')
```
